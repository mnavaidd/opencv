```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
    img = cv2.imread("101.jpg")

    ## convert to hsv
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # Set the orange range
    light_red = (139,0,0) #(170,50,50)
    dark_red = (180,255,255)

    # Apply the red mask 
    mask_red = cv2.inRange(hsv, light_red, dark_red)

    # Set a green range
    light_green = (65,60,60)
    dark_green = (80, 255,255)
    
    # Apply the green mask 
    mask_green = cv2.inRange(hsv, light_green, dark_green)
    
    # Set the yellow range
    light_yellow = (15,0,0)
    dark_yellow = (36, 255, 255)

    # Apply the yellow mask
    mask_yellow = cv2.inRange(hsv, light_yellow, dark_yellow)
    
    # Set the blue range
    light_blue = (149,224,247) #(102,204,255) 
    dark_blue = (102,204,255) #(149,224,247)

    # Apply the blue mask
    mask_blue = cv2.inRange(hsv, light_blue, dark_blue)

    # Combine the four masks
    final_mask = mask_red + mask_green + mask_yellow
    result = cv2.bitwise_and(img, img, mask=final_mask)
    plt.imshow(result)
    
```
